```python
# plot the number of articles per month
import pandas as pd
import matplotlib.pyplot as plt 

#ptt
df = pd.read_csv('../data/ptt_cleaned.csv')
df['date'] = pd.to_datetime(df['date'])

df['month'] = df['date'].dt.to_period('M')
df['year'] = df['date'].dt.to_period('Y')

df = df.groupby(['year', 'month']).size().reset_index(name='count')
df = df.sort_values(by=['year', 'month'])

# dcard
df2 = pd.read_csv('../data/dcard_cleaned.csv')
df2['date'] = pd.to_datetime(df2['date'])

df2['month'] = df2['date'].dt.to_period('M')
df2['year'] = df2['date'].dt.to_period('Y')

df2 = df2.groupby(['year', 'month']).size().reset_index(name='count')
df2 = df2.sort_values(by=['year', 'month'])

# news
df3 = pd.read_csv('../data/news_cleaned.csv')
df3['date'] = pd.to_datetime(df3['date'])

df3['month'] = df3['date'].dt.to_period('M')
df3['year'] = df3['date'].dt.to_period('Y')

df3 = df3.groupby(['year', 'month']).size().reset_index(name='count')
df3 = df3.sort_values(by=['year', 'month'])

# merge by year and month
df = df.merge(df2, on=['year', 'month'], how='outer')
df = df.merge(df3, on=['year', 'month'], how='outer')
df = df.fillna(0)
```

```python
# rename columns
df.columns = ['year', 'month', 'ptt', 'dcard', 'news']

df
```
