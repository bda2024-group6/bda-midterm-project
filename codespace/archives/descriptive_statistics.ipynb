## 查看資料集各公司的佔有比例

```python
import pandas as pd

df_news = pd.read_csv('../data/news_filtered.csv')
stocks_dict = {
    "台船": "2208",
    "長榮": "2603",
    "新興": "2605",
    "裕民": "2606",
    "榮運": "2607",
    "嘉里大榮": "2608",
    "陽明": "2609",
    "華航": "2610",
    "志信": "2611",
    "中航": "2612",
    "中櫃": "2613",
    "萬海": "2615",
    "台航": "2617",
    "長榮航": "2618",
    "亞航": "2630",
    "台灣高鐵": "2633",
    "漢翔": "2634",
    "台驊投控": "2636",
    "慧洋-KY": "2637",
    "宅配通": "2642",
    "長榮航太": "2645",
    "遠雄港": "5607",
    "四維航": "5608",
    "龍德造船": "6753",
    "建新國際": "8367",
    "航運":""
}

# Count the occurrence of each stock in the news title
stock_count = {}
for stock in stocks_dict.keys():
    stock_count[stock] = 0
    for title in df_news['title']:
        if stock in title:
            stock_count[stock] += 1

# Sort the stock_count dictionary by value
stock_count = {k: v for k, v in sorted(stock_count.items(), key=lambda item: item[1], reverse=True)}

# Print the stock_count dictionary
for stock, count in stock_count.items():
    print(f"{stock}: {count}")

# Plot the stock_count dictionary
import matplotlib.pyplot as plt
plt.bar(stock_count.keys(), stock_count.values())
plt.xticks(rotation=90)
plt.rcParams['font.sans-serif'] = ['Noto Sans TC']
plt.rcParams['axes.unicode_minus'] = False
plt.show()



```
